```python
# Q1. Sales by Category Show.
highsales=dmart.groupby('Category')['Sales'].max().reset_index()
highsales
```

```python
# Q2. What is overall sales and profit?
total_sales = dmart['Sales'].sum()
total_profit = dmart['Profit'].sum()

print("Total Sales:", total_sales)
print("Total Profit:", total_profit)
```

```python
# Q3. Top 5 States by Sales?
dmart.groupby('State')['Sales'].sum().sort_values(ascending=False).head(5)
```

```python
# Q4. Year wise Sales?
dmart.groupby('Order_year')['Sales'].sum().plot(kind='bar')
plt.title("Year Wise Sales")
plt.show()
```

```python
# Q5. Profit by Category?
dmart.groupby('Category')['Profit'].sum()
```

```python
# Q6. Sales by Region Show in Pie Chart?
dmart.groupby('Region')['Sales'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title('Region wise Sales',fontsize=26)
plt.show()
```

```python
# Q7. Top 10 Customer?
dmart.groupby('Customer')['Sales'].sum().sort_values(ascending=False).head(10)
```

```python
# Q8. Discount vs Profit Relationship?
dmart.plot.scatter(x='Discount', y='Profit')
plt.show()
```
